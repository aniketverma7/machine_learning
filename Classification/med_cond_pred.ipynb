```python
import pandas as pd

df = pd.read_csv('datasets/medical_conditions_dataset.csv')
df
```

```python
df = df.drop(['id'], axis='columns')
df = df.drop(['full_name'], axis='columns')
df
```

```python
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
df['glucose_levels'] = df['glucose_levels'].fillna(df['glucose_levels'].mean())
df
```

```python
import numpy as np
df1 = df
df1['blood_pressure'] = np.where((df1['condition'] == 'Diabetic') & (df1['blood_pressure'].isna()), 150, df1['blood_pressure'])
```

```python
df1['blood_pressure'] = df1['blood_pressure'].fillna(df1['blood_pressure'].mean())
df1['age'] = df1['age'].fillna(df1['age'].median())
```

```python
df1
```

```python
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()
df['smoking_status_n'] = encoder.fit_transform(df1['smoking_status'])
df['condition_n'] = encoder.fit_transform(df1['condition'])
df['gender_n'] = encoder.fit_transform(df1['gender'])
```

```python
conditon_list = df['condition'].unique()
indx = df['condition_n'].unique()

condition_map = {}

for i in range(len(conditon_list)):
    condition_map[int(indx[i])] = conditon_list[i]

condition_map
```

```python
df1
```

```python
df2 = df1.drop(['smoking_status', 'condition', 'gender'], axis='columns')
df2
```

```python
X = df2.drop(['condition_n'], axis='columns')
Y = df2['condition_n'] 
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_jobs=20)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
